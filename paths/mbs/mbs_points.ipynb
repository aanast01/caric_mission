```python
import numpy as np

p1 = [ [0  ,  8  ],
       [3  ,  25 ],
       [26 ,  18 ],
       [22 ,  2  ],
     ]

p2 = [ [31 ,  0  ],
       [36 ,  17 ],
       [58 ,  8  ],
       [48 , -9  ],
     ]

p3 = [ [58 , -13 ],
       [80 , -26 ],
       [87 , -10 ],
       [67 , 4.5 ],
     ]

p4 = [ [0  , 4   ],
       [25 , 1   ],
       [30 , 17  ],
       [58 , 9   ],
       [54 , -16 ],
       [80 , -30 ],
       [54 , -16 ],
       [58 , 9   ],
       [30 , 17  ],
       [25 , 1   ],
       [0  , 4   ],
     ]

p5 = [ [-2 ,  25  ],
       [ 30 , 18  ],
       [ 25 , 1   ],
       [ 53 , -11 ],
       [ 64 , 8   ],
       [ 93 , -10 ],
       [ 64 , 8   ],
       [ 53 , -11 ],
       [ 25 , 1   ],
       [ 30 , 18  ],
       [-2 ,  25  ],
     ]

p1 = np.array(p1)
p1 = np.transpose(p1)
p2 = np.array(p2)
p2 = np.transpose(p2)
p3 = np.array(p3)
p3 = np.transpose(p3)
p4 = np.array(p4)
p4 = np.transpose(p4)
p5 = np.array(p5)
p5 = np.transpose(p5)


def modify(p, h, loop):
    # p = p - p[:, 0, None]
    p_ = p
    for n in range(0, loop):
        p_ = np.hstack([p_, p])
    p_h = np.vstack([p_, h*np.ones([1, p_.shape[1]])])
    return p_h


p1_h = modify(p1, 30, 5)
p2_h = modify(p2, 40, 5)
p3_h = modify(p3, 50, 5)
p4_h = modify(p4, 20, 2)
p5_h = modify(p5, 25, 2)
```

```python
np.savetxt('path1.txt', np.vstack([p1_h, np.zeros([1, p1_h.shape[1]])]), delimiter=',', fmt='%3.0f')
np.savetxt('path2.txt', np.vstack([p2_h, np.zeros([1, p2_h.shape[1]])]), delimiter=',', fmt='%3.0f')
np.savetxt('path3.txt', np.vstack([p3_h, np.zeros([1, p3_h.shape[1]])]), delimiter=',', fmt='%3.0f')
np.savetxt('path4.txt', np.vstack([p4_h, np.zeros([1, p4_h.shape[1]])]), delimiter=',', fmt='%3.0f')
np.savetxt('path5.txt', np.vstack([p5_h, np.zeros([1, p5_h.shape[1]])]), delimiter=',', fmt='%3.0f')
```
